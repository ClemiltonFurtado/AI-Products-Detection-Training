```python
import torch
from ultralytics import YOLO

print("CUDA available?", torch.cuda.is_available())
print("GPU Name:", torch.cuda.get_device_name(0))

model = YOLO("yolov8n.pt")
model.to("cuda")

print("YOLO running in your GPU!")
```

# Train the model using the SKU-110K

```python
results = model.train(data="SKU-110K.yaml", 
                      epochs= 10, 
                      imgsz= 640, 
                      batch = 16,
                      device = 'cuda', 
                      )
```
